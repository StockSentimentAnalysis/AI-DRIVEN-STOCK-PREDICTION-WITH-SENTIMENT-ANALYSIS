```python
import pandas as pd
import numpy as np

# --------------------------------------------------
# Load your existing results
# --------------------------------------------------

INPUT_CSV = "Annual_report_finbert_absa.csv"
OUTPUT_CSV = "Annual_report_finbert_absa_normalized.csv"

df = pd.read_csv(INPUT_CSV)

# --------------------------------------------------
# Identify aspect columns
# --------------------------------------------------

aspect_cols = [
    "profit_sentiment",
    "deposits_sentiment",
    "advances_sentiment",
    "capital_sentiment",
    "npl_sentiment",
    "provisions_sentiment",
    "risk_sentiment",
    "outlook_sentiment",
    "macroeconomic_sentiment",
    "costs_sentiment"
]

# --------------------------------------------------
# Z-score normalize EACH aspect across years
# --------------------------------------------------

for col in aspect_cols:
    mean = df[col].mean()
    std = df[col].std(ddof=0)

    df[col + "_z"] = (df[col] - mean) / std

# --------------------------------------------------
# Recompute calibrated WASI
# --------------------------------------------------

z_cols = [c + "_z" for c in aspect_cols]

df["overall_wasi_z"] = df[z_cols].mean(axis=1)

# --------------------------------------------------
# Optional: squash extreme values (recommended)
# --------------------------------------------------

df["overall_wasi_z"] = df["overall_wasi_z"].clip(-3, 3)

# --------------------------------------------------
# Save calibrated output
# --------------------------------------------------

df.to_csv(OUTPUT_CSV, index=False)

print(f"Normalized & calibrated results saved to: {OUTPUT_CSV}")
```
